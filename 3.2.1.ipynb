## 3.2.1字符串
&emsp;&emsp;在Python语言中，字符串是最常用的数据类型之一。可以把一个或多个字符用单引号（'）、双引号（"）或者三引号（'''）括起来，三者的差别在于：使用单引号表示字符串时，字符串中可以包含双引号；使用双引号表示字符串时，字符串中可以包含单引号；使用三引号表示字符串时，字符串内容可以包含单引号、双引号和换行符。

### 1. 字符串创建
&emsp;&emsp;**例3-2-1** 字符串的常量表示及转义字符使用示例

```python
print("It's fine!")
```

```python
print ("hello everyone\ntoday is a great day!")
```

&emsp;&emsp;Python同样支持以"\"为前缀的转义字符，例如使用转义字符"\n"可以在输出时使字符串换行。只有引号''或""无任何字符的为空字符串。

&emsp;&emsp;**例3-2-2** 字符串变量示例

```python
s="hello"
```

```python
print(s)
```

```python
t=s
```

```python
id(s)==id(t)    #t和s指向同一个字符串对象"hello"
```

&emsp;&emsp;字符串同样也可以使用字符串变量来操作。字符串变量的实质是一个指向字符串对象的标识符。
### 2. 字符串的基本运算
&emsp;&emsp;Python中提供了5个字符串的基本运算符。


#### （1）索引操作
&emsp;&emsp;字符串是一个序列对象，每一个字符都有索引号，也称为下标。而且Python支持正向索引号和反向索引号两种索引体系。

&emsp;&emsp;使用下标来获取字符串中指定的某个字符，称为索引操作，下标是一个整数值，可以是整数常量，整数变量，也可以是一个整数表达式，用法如下：

&emsp;&emsp;&emsp;&emsp;<字符串>[下标]

&emsp;&emsp;**例3-2-3** 字符串索引操作示例。

```python
s="hello Python!"
```

```python
s[0]
```

```python
i=10
```

```python
s[i+1]
```

```python
s[-1]
```

&emsp;&emsp;注意：Python中下标位置是从0开始计数的，数值表达式可以为负数，则表示从右向左计数。

#### （2）切片操作
&emsp;&emsp;Python提供了切片操作获取子串，子串是一个字符串中连续或不连续的部分字符。常用的使用方法为：

&emsp;&emsp;&emsp;&emsp;<字符串>[ start:end:step]

&emsp;&emsp;即获取索引号从start到end-1的间隔为step的子串。

&emsp;&emsp;**例3-2-4** 获取字符串的子串示例

```python
str="Python语言"
```

```python
str[:6]    #截取str字符串下标从0到5的子串
```

```python
str[6:]    #截取str字符串下标从6到最后的子串
```

```python
str[-2:]   #使用逆序下标，-2与6是一致的
```

```python
s='abcdefg'
```

```python
print(s[::])  #获取原串
```

```python
print(s[::-1])  #翻转字符串
```

#### （3）连接和复制操作
&emsp;&emsp;字符串类型支持使用“+”联接两个字符串，使用“*”进行字符串的复制。

&emsp;&emsp;**例3-2-5** 联接运算示例

```python
s= 'shang'
```

```python
s+='hai'
```

```python
s
```

&emsp;&emsp;**例3-2-6** 复制运算示例

```python
s="hi"
```

```python
t=s*3
```

```python
t
```

#### （4）子串测试运算
&emsp;&emsp;子串测试运算in可以测试一个子串是否存在于另一个字符串中，计算结果返回布尔值。用法为：

&emsp;&emsp;&emsp;&emsp;<子串> in<字符串>

&emsp;&emsp;**例3-2-7** 子串测试运算示例

```python
s='python'
```

```python
'py' in s
```

```python
t='the'
```

```python
t in s
```

#### （5）str对象的方法
&emsp;&emsp;str类型提供了丰富的字符串操作的方法，如格式化字符串方法format()即是str对象的方法之一。表3-2-2所示为常用str对象方法， S表示一个str对象。读者同样可以通过help(str)查询更多的字符串操作的方法。

&emsp;&emsp;str对象方法的调用形式为：

&emsp;&emsp;&emsp;&emsp;<字符串>.方法名(<参数>)

&emsp;&emsp;在交互方式的提示符>>>后输入 对象名. 稍作停留，会显示该对象的所有方法的列表，使用上下光标键可以选择所需的方法。

&emsp;&emsp;**例3-2-8** str对象方法示例

```python
s='   hello python   '
```

```python
t=s.strip()   #去除左右空格
```

```python
s
```

```python
t
```

```python
t.upper()      #所有字符转换成大写
```

```python
t.count('o')    #统计查找子串'o'出现的次数
```

```python
t.find('o')     #寻找子串'o'，返回首次出现的下标值
```

```python
t.replace('hello','hi')  #字符串'hello'替换成'hi'字符串
```
