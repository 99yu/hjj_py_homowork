## 3.4.1 模块
&emsp;&emsp;一般的高级语言程序系统中都提供系统函数丰富语言功能。Python的系统函数由标准库中的很多模块提供。标准库中的模块，又分成内置模块和非内置模块，内置模块_builtin_中的函数和变量可以直接使用，非内置模块要先导入模块，再使用。
### 1. 内置模块
&emsp;&emsp;Python中的内置函数是通过_builtin_模块提供的，该模块不需手动导入，启动Python时系统会自动导入，任何程序都可以直接使用它们。该模块定义了一些软件开发中常用的函数，这些函数实现了数据类型的转换，数据的计算，序列的处理、常用字符串处理等等，常见内置函数如表3-4-1所示。


&emsp;&emsp;内置函数的调用方式与数学函数类似，函数名加上相应的参数值，多个参数值之间以逗号分隔：

&emsp;&emsp;&emsp;&emsp;<函数名>（参数序列）

&emsp;&emsp;**例3-4-1** 内置模块函数示例

```python
round(78.3456,2)  #按四舍五入对78.3456保留2位小数
```

```python
len("Good morning")  #使用计算字符串的长度
```

```python
pow(3.0,2) #计算3.0的3次方
```

### 2. 非内置模块
&emsp;&emsp;非内置模块指的是Python的一些标准库。如：数学运算的math库、random库、decimal库，日期和时间处理的datetime库、time库库，文字处理的re库，数据对象的array库、queue库、系统信息的sys库、os库和网络应用的socket库等。标准库函数支持程序功能的快速开发。
#### （1）库的导入和使用方法1
&emsp;&emsp;标准库函数在使用前要先导入其所在模块库，Python中使用import语句来导入标准库：

&emsp;&emsp;&emsp;&emsp;import  <模块名>

&emsp;&emsp;其中模块名也可以有多个，多个模块之间用逗号分隔。该语句通常放在程序的开始部分。模块导入后，可以在程序中使用模块中定义的函数或常量值。方法如下：

&emsp;&emsp;&emsp;&emsp;<模块名>.<函数>(<参数>)

&emsp;&emsp;&emsp;&emsp;<模块名>.<字面常量>

&emsp;&emsp;**例3-4-2** math数学库使用示例1

```python
import math        #导入数学库
```

```python
math.pi            #查看圆周率π常数，调用时需要math模块名作为前缀
```

```python
math.pow(math.pi,2)    #函数math.pow(x,y)即为求x的y次方
```

```python
#计算边长为8.3和10.58，两边夹角为37度的三角形的面积的表达式
8.3*10.58*math.sin(37.0/180* math.pi)/2
```

#### （2）库的导入和使用方法2
&emsp;&emsp;还可以通过import命令明确引入模块的函数名，有多个函数导入时函数名之间用逗号分隔：

&emsp;&emsp;&emsp;&emsp;from <模块名> import <函数名1>,<函数名2>…

&emsp;&emsp;使用这种方法导入指定的函数，调用时函数名前不需要加模块前缀，方法如下：

&emsp;&emsp;&emsp;&emsp;<函数>(<参数>)

&emsp;&emsp;或者通过星号（*）导入模块中的所有函数或其他对象：

&emsp;&emsp;&emsp;&emsp;from <模块名> import *

&emsp;&emsp;调用时同样在函数名前不需要加模块前缀，但要注意所导入模块中的函数名等与现有系统中的函数不能产生冲突。

&emsp;&emsp;**例3-4-3** math数学库使用示例2

```python
from math import sqrt     #只导入数学库中的sqrt()函数
```

```python
sqrt(16)                  #调用sqrt()无需math模块名前缀
```

```python
from math import *       #导入数学库中所有的函数及其他对象
```

```python
sqrt(16)
```
