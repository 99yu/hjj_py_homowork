```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 
x = np.arange(0, 2 * np.pi, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#RC
plt.rcParams["lines.linestyle"]="-."      #
plt.rcParams["lines.linewidth"]=3         #
```

```python
# 
plt.title('y = sin(x) and y=cos(x)')
plt.xlabel('x')
plt.ylabel('y')

#
plt.plot(x, y1,"r")                       #
plt.plot(x, y2,"k")                       #
plt.grid()
plt.show()
```
