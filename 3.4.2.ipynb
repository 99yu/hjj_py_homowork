## 3.4.2 自定义函数
### 1. 函数的定义
&emsp;&emsp;函数的定义包括函数名称、形参以及函数体，定义函数的语法如下：

&emsp;&emsp;&emsp;&emsp;def  函数名(形参列表)：

&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;函数体

&emsp;&emsp;**例3-4-4** 求最大公约数的gcd()函数定义。如图3-4-1所示求两个整数的最大公约数的gcd()函数定义。

&emsp;&emsp;gcd()函数定义包括函数头和函数体。

* 函数头

&emsp;&emsp;从关键字def开始，接着是函数名，函数名命名应该体现“望名生义”，此处gcd是最大公约数（Greatest Common Divisor）的缩写。函数名后的一对圆括号是函数的标志符，括号中定义参数，多个参数间以逗号分隔。函数定义处的参数称为形式参数，简称形参。

* 函数体

&emsp;&emsp;是实现函数功能的程序语句段。先找到x、y两个整数变量中较小的一个赋值给smaller变量，遍历i从1~smaller，验证i是否是x和y的公约数，如果是则记录在变量g中。在遍历的过程中，找到一个公约数，就改写g，最后一个就是最大公约数。最后的return语句，将计算的最大公约数作为返回值返回到调用处。

&emsp;&emsp;函数区分为有返回值的函数和无返回值的函数。有返回值的函数需要通过return语句返回一个值，无返回值的函数体中可以无return语句。

&emsp;&emsp;执行到return语句意味着函数执行的终止。同样无return语句的函数当函数体执行结束时也会终止该函数的执行。

### 2. 函数的调用
&emsp;&emsp;函数定义后，未经调用，是不会执行的。使用函数需要使用函数调用语句。在函数调用的时候要求传递对应的值给形参，这个传递的值称为实际参数，简称实参。参数是可选的，也就是说函数可以不包括参数，例如：random.random() 就不包含参数。

&emsp;&emsp;根据函数是否有返回值，函数的调用也是不同的。
#### （1）有返回值的函数调用
&emsp;&emsp;如果函数有返回值，函数调用的时候需要安排接受返回值。

* 可以是赋值语句，将返回值赋值给一个变量

&emsp;&emsp;例如：如图3-4-1所示的函数调用z=gcd( 28,70)  

&emsp;&emsp;求28和70的最大公约数，28和70是实参，28传递给形参x，70传递给形参y。函数执行结束，return语句将g的值返回到调用处，赋值给变量z，执行后z的值为14。  

* 可以是表达式语句，将返回值作为表达式的一个数值，继续参加运算

&emsp;&emsp;例如：print(gcd(28,70)*gcd(26,65))

&emsp;&emsp;输出28和70的最大公约数与26和65的最大公约数的乘积。返回的最大公约数作为表达式的一部分继续参加乘法运算。

* 可以作为函数调用的实参

&emsp;&emsp;例如：z=gcd(gcd(28,70),21)

&emsp;&emsp;gcd(28,70)的返回值14作为外层gcd()函数的第一个实参，和21一起求最大公约数，执行后z的值为7。

#### （2）无返回值的函数调用
&emsp;&emsp;print()函数的调用是一个典型的无返回值的函数调用，执行一个输出的操作。没有返回值的函数的调用形式是函数语句。

&emsp;&emsp;例如：print( num1,"和", num2,"的最大公约数为", gcd(num1, num2))

&emsp;&emsp;函数语句就是一个函数调用，从函数名开始，一对圆括号，圆括号中是实参列表。

&emsp;&emsp;**例3-4-5** pause()函数的定义及调用

#pause()函数的定义

```python
def pause(times):
    c=0
    while(c< times):
        c=c+1    
```

#pause()函数的调用

```python
pause(100000000)
```

&emsp;&emsp;说明：pause()函数的作用是暂停，时间长短由参数times的值决定。

#### （3）函数的执行过程
&emsp;&emsp;当程序调用一个函数是，程序的控制权就会转移到被调用的函数，当执行到函数结束或执行到一个return语句，函数将程序的控制权归还给函数调用处。

&emsp;&emsp;**例3-4-6** 完整程序，键盘输入两个整数，调用gcd()函数求两个整数的最大公约数。

&emsp;&emsp;程序完整实现代码如下：

```python
def gcd(x, y):
    smaller=x if x<=y else y
    for i in range(1,smaller + 1):
       if((x % i == 0) and (y % i == 0)):
           g = i
    return g
```

```python
def main():
    num1 = int(input("输入第一个数: "))
    num2 = int(input("输入第二个数: "))
    print( num1,"和", num2,"的最大公约数为", gcd(num1, num2))
```

```python
main()
```

&emsp;&emsp;这个程序包含了gcd()函数和main()函数，在main()函数的print()函数的实参位置调用了gcd()函数。执行过程如图3-4-2所示：

&emsp;&emsp;Python翻译器从程序的第一行开始逐行读取脚本，当它读取gcd()函数和main()函数的函数定义时，将函数存储在内存中并不执行。最后读取到main()函数调用语句，程序开始执行，调用main()函数，程序的控制权转向main()函数，如箭头①所示。

&emsp;&emsp;main()函数的执行从函数的第一句开始，先输入两个数给num1和num2变量，然后执行print()函数，这时调用gcd(num1,num2)，程序的控制权转移到gcd()函数，如箭头②所示，开始执行gcd()函数。

&emsp;&emsp;当gcd()函数执行到return语句，程序的控制权返回到main()函数的调用处，如箭头③所示，完成print()函数的输出。此时main()函数的语句全部执行完毕，程序的控制权返回到主程序的调用处，如箭头④所示，整个程序结束。

&emsp;&emsp;在Python中，所有的函数定义在执行时，都被Python翻译器读入到内存，函数定义的顺序和函数的执行顺序是无关的。也就是说，多个函数定义的顺序是可前可后的，只需函数的调用语句出现在函数定义之后即可。
